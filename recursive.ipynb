```python
def runCommand(cwd):
    for (dirs,dirname,files) in os.walk("."):
        for file in files:
            if "epsilonAA" in dirs and "input.lmp" in file and "5000" not in dirs and "epsilonAA_100" not in dirs:
                f = dirs
                #os.system("cp rerun_input.lmp {}".format(f))
                os.system("cp chain_dist_from_patch.py {}".format(f))
                os.system("cp rerun_input.lmp {}".format(f))
                os.chdir(dirs)
                os.system("cat eq_1.xyz eq_2.xyz prod.xyz > full.xyz")
                os.system("lmp_serial -in rerun_input.lmp")
                os.system("python chain_dist_from_patch.py")
                os.chdir(cwd)
                
                print(f)
```

```python
if __name__ == '__main__':
    cwd = os.getcwd()
    runCommand(cwd)
```
